```python
import pandas as pd
```

```python
# read in data
full_veh = pd.read_csv('Vehicle__Snowmobile__and_Boat_Registrations.csv', skipinitialspace=True)
```

```python
# remove non passenger vehicles
pas_veh = full_veh.loc[full_veh['Registration Class'].isin(['PAS', 'SRF', 'MOT', 'HSM', 'LMA', 'LMB', 'LMC'])]

# remove extra columns
pas_veh = pas_veh[['Record Type', 'VIN', 'Registration Class', 'City', 'State', 'Zip', 'County', 
                   'Model Year', 'Make', 'Body Type', 'Fuel Type']]
```

```python
# kept NY, CT, NJ, PA
pas_veh = pas_veh.loc[pas_veh['State'].isin(['NY', 'CT', 'NJ', 'PA'])]
#pas_veh['County'] = pas_veh['County'].str.strip()
```

```python
# Strip white space
for i in range(0, len(pas_veh.columns) - 1):
    if(type(pas_veh.iloc[:,i][0]) == 'str'):
        pas_veh.iloc[:,i] = pas_veh.iloc[:,i].str.strip()
```

```python
# remove upstate NY, keep any of NJ, CT or PA plates
improved_ny = pas_veh.loc[(pas_veh['County'].isin(['ORANGE', 'PUTNAM', 'ROCKLAND', 'WEST CHESTER',
                                                          'BRONX', 'NEW YORK','RICHMOND', 'KINGS','QUEENS',
                                                          'NASSAU', 'SUFFOLK', 'DUTCHESS', 'ULSTER', 'SULLIVAN']))
                           & (pas_veh['State'] == 'NY')
                         ]
```

```python
# drop old NYC rows
pas_veh = pas_veh[pas_veh['State'] != 'NY']
# add back in filtered
pas_veh = pas_veh.append(improved_ny)
```
